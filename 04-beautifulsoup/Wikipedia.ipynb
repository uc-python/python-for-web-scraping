Some information we might want
- Page title (string)
- Image
- All body text (string)

```python
import requests
from bs4 import BeautifulSoup
```

```python
# url = 'https://en.wikipedia.org/wiki/Sherlock_Holmes'
# url = 'https://en.wikipedia.org/wiki/Elizabeth_Bennet'
url = 'https://en.wikipedia.org/wiki/Bilbo_Baggins'
# filename = 'sherlock.txt'
# filename = 'elizabeth.txt'
filename = 'bilbo.txt'
response = requests.get(url)
response.content
```

```python
bs = BeautifulSoup(response.content)
```

### Page Title

```python
title = bs.title.string
```

### Image

```python
infobox = bs.find(name='table', class_='infobox')
image = infobox.img
image
```

```python
image_url = image['src']
```

### All Text

```python
all_paragraphs = bs.find(id='mw-content-text').find_all('p')
```

```python
p_text = ''
for paragraph in all_paragraphs:
    # Append the concatenated text of the newest paragraph to p_text
    p_text = p_text + ''.join(paragraph.strings)
    p_text = p_text + '\n'
    
print(p_text)
```

### Save File

```python
with open(filename, 'w') as f:
    f.write('Title: ')
    f.write(title)
    f.write('\n')
    f.write('Image URL: ')
    f.write(image_url)
    f.write('\n')
    f.write(p_text)
print('Done writing!')
```
