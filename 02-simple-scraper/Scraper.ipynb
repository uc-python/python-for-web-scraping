```python
URL = 'https://www.sec.gov/Archives/edgar/data/56873/000155837021003706/kr-20210130x10k.htm'
PATTERN = 'Common shares'
```

```python
import requests
import pandas as pd
```

```python
response = requests.get(URL)
```

```python
result = pd.read_html(response.content, match=PATTERN)
```

```python
df = result[0]
```

```python
df
```

```python
new_columns = ['label', 'drop_1', 'drop_2', 'fy_2020', 'drop_3', 'drop_4', 'fy_2019', 'drop_5']
df.columns = new_columns
```

```python
df
```

```python
df = df.drop(['drop_1', 'drop_2', 'drop_3', 'drop_4', 'drop_5'], axis=1)
```

```python
df = df.drop([0, 1, 2], axis=0)
```

```python
df.head(10)
```

```python
df = df.set_index('label')
```

```python
df.loc['Receivables':'LIFO reserve', ['fy_2020']]
```
